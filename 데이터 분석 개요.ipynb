# 주제 : 지역경제 활성화를 위한 여행 데이터분석 및 워케이션 지역선정 및 특화 전략

# 1.  프로젝트를 들어가며


### **분석 배경**
##### 1. 분석 목적
- 코로나 19 이후 관광인구가 해외로 유출되고 있어 사람들이 국내여행을 오게끔 할수있는 방법을 모색 -> 장기적으로 머물고 평일에도 체류인구를 증가시킬수 있는 워케이션을 분석대상으로 선정
- 워케이션의 여행특성 분석을 통한 새로운 워케이션 상품 개발
- 워케이션 선호 지역중 관광 활성화가 부족한 지역을 대상으로 지자체와 협력해 워케이션 관광객 유치를 통한 지역경제 활성화를 도모

##### 2. 분석 내용
- 워케이션 선호지역의 활성화요인 파악
- 여행 형태 분석(숙박 형태, 관광목적, 장소, 체류기간, 접근성(액세서빌리티) 등)
- 지역경제활성화를 위한 워케이션 적정 지역 선정을 위한 방문자, 관광소비액 등 분석

##### 3. 워케이션의 개념과 현황
- 워케이션(Workation)은 '일(Work)'과 '휴가(Vacation)'의 합성어로, 업무와 휴식을 동시에 추구하는 새로운 근무 형태 이자 새로운 여행 트렌드 입니다. 
- 코로나19 이후 원격근무가 보편화되면서 워케이션에 대한 관심이 증가하고 있습니다.
- 디지털 노마드 개념의 등장과 새로운 근무 형태의 확산
- 워케이션의 지방소멸 대응책으로서의 부상

# Phase3. 데이터 분석

### 추가정제
- 기준년월 열을 문자열로 변환
- 이름이 중복되는 지역구 확인후 추가 정제

### 데이터프레임 분리
- 지역구별 데이터프레임 분리
- 워케이션 언급량 상위 30% 지역/ 하위 30% 지역 데이터프레임 분리
- 각 년도별 지역구 데이터프레임 분리

### 기본기술통계 확인
- 방문자수 : 최소 3.7백만명의 방문자수를 보이고 4사분위수 구간을 살펴보았을때 25%, 50%, 75% 구간에서 각각 고른 분포를 보였다. 
            그러나 75% ~ 100% 구간 사이의 분포가 넓고 특정 지역에 방문자수가 밀집되어있음을 알 수 있다.

- 방문자수증감률 : 90% ~ 100% 구간에 분포가 밀집되어있다. 따라서 일부 지역의 증감률이 매우 높음을 알 수 있다. 

- 관광소비액 : 평균보다 표준편차가 크므로 관광소비액이 매우 넓은 분포를 보임을 알 수 있다. 방문자수와 마찬가지로 90% ~ 100%의 일부 구간에 분포가 밀집되어있음을 알 수 있다. 또한 0%구간간에서 100%구간으로 갈수록 분포가 점점 넓어지는 특성이 있다.

- 체류시간 : 25%이하 구간, 75%이상 구간에서 밀집된 분포를 보이며 이 또한 특정 지역의 체류시간이 극단적으로 낮거나 높음을 나타낸다.

- 숙박방문자비율 : 평균과 표준편차를 보았을때는 안정적이나 75%이상 구간에서 100%구간까지 표준편차의 3배넘는 차이의 분포를 보인다. 이는 특정 지역의 숙박방문자비율이 매우 높음을 나타낸다. 

- 평균 숙박일수 : 대부분의 분포가 평균 2.8 에서 멀리 떨어지지 않은 평균 가까이 밀집되어 있는 분포를 보인다. 최소 , 최대 값의 차이가 대략 1일차이로 평균 숙박일수는 대체적으로 안정적인 분포를 보인다.

- 단기숙박비율 : 평균 88, 표준편차 3으로 대체적으로 안정적인 분포를 보인다. 다만 90%이상 구간에서 다소 넓은 분포를 보이는 특징이 있다. 따라서 특정 상위 10%지역에 단기숙박비율이 높게 분포되어 있는 것으로 보인다.

- 장기숙박비율 : 평균 11, 표준편차 3으로 대체적으로 안정적인 분포를 보인다. 다만 10%이하 구간에서 다소 넓은 분포를 보이는 특징이 있다. 따라서 특정 하위 10%지역에 장기숙박비율이 낮게 분포되어 있는 것으로 보인다. 

- 전년대비단기숙박증가률 : 10%이상 구간 ~ 90%이하 구간에서는 각 표준편차 만큼의 적절한 분포를 보인다. 그러나 10%이하, 90%이상 구간에서는 다소 넓은 분포를 보이는 특징이 있다. 따라서 특정지역의 단기숙박비율이 감소하거나 증가하는 경향이 있다. 

- 전년대비장기숙박증가률 : 전년대비장기숙박증가률과 같이 10%이상 구간 ~ 90%이하 구간에서는 각 표준편차 만큼의 적절한 분포를 보인다. 그러나 10%이하, 90%이상 구간에서는 다소 넓은 분포를 보이는 특징이 있다. 따라서 특정지역의 장기숙박비율이 감소하거나 증가하는 경향이 있다. 

- 워케이션 언급량 : 평균이 21이고 표준편차가 30으로 데이터가 평균에서 멀리 떨어져 분포되어있음을 알 수 있다. 50%구간에서 100%구간으로 갈수록 점점 수치가 커지면서 분포가 넓어지는 경향이 있다. 따라서 특정 중상위 지역에 워케이션 언급량이 많은 것을 알 수 있다.

### 워케이션 언급량 분석
- 연도별 워케이션 언급량 및 성장률 변화 분석

    - 워케이션 언급량이 년도를 거듭할수록 꾸준히 상승중이다. 
    - 그러나 성장세는 줄어드는 추세이다. 
    - 2021년에서 2022년까지의 워케이션 언급량의 성장세는 약 124%로 워케이션 관심도가 가장 컸던 시기임을 알 수 있다.

- 지역시별 워케이션 언급량 비율 분석
    - 지역시별 워케이션 언급량 분포는 강원도 > 서울 = 부산 > 제주 > 경기 순으로 높은 수치를 나타낸다.
    - 워케이션 사업을 처음 추진한 강원도가 누적된 워케이션 언급량이 가장 많음을 알 수 있다.

- 각 지역시의 워케이션 성장률 분석
    - 부산 워케이션 언급량이 2021년 대비 2024년 933% 증가한 것으로 나타남 (이유)
    - 2023년 대비 2024년 증감률은 27%로 가장 높은 성장률을 나타냄.
    - 다른 지역 시/도 에 비해 뚜렷한 증가세를 보임.
    - 강원도와 부산은 최근까지도 어느정도의 성장률을 보이며 상승추세이다.
    - 경기도,서울,제주는 2021년 대비 2024년 증감률에 비해 최근 2023년대비 2024년 증감률이 낮은 수치를 보이며 다소 성장세가 주춤하는 추세이다.
    - 특히 제주도는 2021년 대비 2024년 증감률에 비해 최근 2023년대비 2024년 증감률이 상대적으로 매우 떨어져 큰 차이를 보인다.

- 연도별 워케이션 언급량 상위 지역 TOP10
    - 2021년 강원도 지역구의 워케이션 언급량이 높은 수준을 유지하였지만 2024년 제주도 지역구의 언급량이 상위 2위 모두를 차지했다.
    - 서귀포시의 워케이션 언급량이 2024년에 급하락하였다.



### 지역별 주요 관광요소 분석
- 지역시별 평균 방문자수 분석
    - 각년도별 전체 지역구의 '방문자수', '관광소비액(천원)', '단기숙박비율', '장기숙박비율' 상관관계를 보았을때는 '단기숙박비율'에 따라 '방문자수', '관광소비액(천원)'이 증가하는 패턴이다.
    - 각년도별 워케이션 언급량 상위 5개 지역구의 '방문자수', '관광소비액(천원)', '단기숙박비율', '장기숙박비율' 상관관계를 보았을때는 '장기숙박비율'에 따라 '방문자수', '관광소비액(천원)'이 증가하는 패턴이다.
    - 전체 지역구로 볼땐 단기 숙박 비율이 높았지만 워케이션 지역 한정 장기숙박비율이 더 높다고 판단할수있다.

- 지역별 주요 관광요소 분석
    - 워케이션 상위 30%지역의 방문자수, 관광소비액, 숙박방문자 비율이 워케이션 하위 30%지역보다 높게 나타난다.
    - 그러나 워케이션 상위 30%지역이 워케이션 하위 30%지역보다 단기숙박비율이 높고 장기숙박비율이 낮게 나타난다.

- 주요 관광요소간의 상관관계
    - 단기숙박비율, 숙박방문자비율 간의 높은 상관관계를 보인다.
    - 숙박방문자비율, 체류시간 간의 높은 상관관계를 보인다.
    - 장기숙박비율, 평균숙박일수 간의 높은 상관관계를 보인다. <br>
    <해석>
    - 숙박방문자비율과 체류시간은 상관관계가 있다. 그러나 숙박일수와 체류시간은 상관관계가 없다.
    - 숙박방문자비율과 단기숙박비율은 상관관계가 있다. 그러나 숙박방문자비율과 장기숙박비율은 상관관계가 없다.
    - 단기숙박비율, 숙박방문자비율은 워케이션 언급량과 약한 양의 상관관계가 있다.
    - 단기숙박하는 사람이 많다. 따라서 숙박방문자가 많아진다.
    - 장기숙박하는 사람이 적다.
    - 단기숙박비율이 늘수록 워케이션 언급량이 늘어난다.

    - "단기숙박하는 방문자의 비율이 높을수록 전체 숙박방문자 비율도 높아지는 경향이 있으며, 이는 워케이션 언급량과도 약한 양의 상관관계를 보인다. 반면, 장기숙박은 전체 숙박방문자 패턴에 유의미한 영향을 미치지 않는다. 체류시간은 숙박방문자 비율과는 관련이 있으나, 실제 숙박일수와는 독립적인 패턴을 보인다."
    - 관광소비액은 워케이션 언급량, 숙박일수, 장기숙박비율 과 약한 양의 상관관계를 가진다. 
    - 이는 다른 변수인 숙박방문자 비율, 단기숙박비율, 체류시간 보다 관광소비액에 더 큰 영향을 미치는 것으로 보인다.
    - 따라서 장기숙박 성격의 워케이션 활성화가 지역관광활성화에 영향을 미치는 것으로 해석할 수 있다.

- 장기숙박비율과 관광소비액의 산점도
    - 장기숙박비율에 따라 관관소비액이 선형 증가하는 경향을 보기는 어렵지만 일부 지역에서 장기숙박비율이 상대적으로 큰 구간에서 관광소비액이 높은 수치를 나타내는 것을 알 수 있다. 

- 지역별 장기숙박비율 비율과 관광소비액의 상관관계 분석
    - 워케이션 상위 30%지역의 장기숙박비율과 관광소비액간의 상관관계가 양의 상관관계로 유의미한 상관관계를 나타낸다.

- 워케이션 언급량 상위 5개 지역구의 장기숙박비율에 따른 방문자수, 관광소비액 비율 분석
    - 워케이션 언급량 상위 5개 지역의 모든년도에서 방문자수,관광소비액,장기숙박비율이 모두 비례한다.
    - 워케이션 언급량 상위 5개 지역의 모든년도에서 방문자수와 장기숙박비율이 강한 양의 상관관계를 갖는다.

- 네비게이션 목적지 언급량 상관관계
    - 2021년에는 쇼핑과 음식이 방문자수, 관광소비액에 미치는 영향이 컸지만
    - 2024년에는 음식의 영향력이 약해지고 쇼핑,관광,숙박의 영향력이 커졌다.
    - 가장 주목할만한 변화는 2022년부터 자연관광-워케이션 언급량의 상관관계가 높은 수준을 유지하고 있다는 점이다.
    -  2024년 기준 숙박,워케이션 언급량 / 자연관광,워케이션 언급량 간의 0.5이상의 양의 상관관계를 보인다.

- 소셜미디어 트렌드 언급량 상관관계
    - 각각의 소셜미디어 트렌드 언급량과 워케이션 언급량과의 상관관계가 년도를 지날수록 점점 커진다.
    - '기타' 트렌드 언급량과 워케이션 언급량과의 상관관계가가 년도를 지날수록 점점 커진다.
    - 2024년 '체험' 과 '기타' 트렌드 언급량과 워케이션간의 상관관계가 가장 높게 나타났다.

    - 이는 현재 워케이션이 활성화된 지역에 장기관광 중심의 소비 패턴이 정착되어 있음을 의미
    - 워케이션으로 지역경제활성화하는 이유가 장기숙박인구 유치해 지역소비촉진
    - 1차선정 지역모두 장기관광 중심의 소비패턴이 약하다. 
    - 각 지역의 시계열 데이터에서 장기숙박비율과 관광소비액의 상관관계가 개선되는 지역과 악화되는 지역을 나눔
    - 장기숙박비율과 관광소비액의 상관관계가 개선되는 지역에 장기관광 중심의 워케이션 사업을 추진할 마케팅 전략이 필요

# Phase.5 워케이션 콘텐츠 마케팅 전략 보고서
## 1. 개요
- 최근 워케이션(Workation) 트렌드는 지속적으로 상승하고 있으며, 특정 지역에서의 언급량 변화 및 관광 요소와의 상관관계를 통해 효과적인 마케팅 전략을 도출할 수 있습니다. 본 보고서는 분석 결과를 바탕으로 워케이션 콘텐츠 마케팅 전략을 수립하고, STP 전략, 4P 전략, 3C 전략을 정리하여 제안합니다.
## 2. 워케이션 트렌드 분석
1. 워케이션 언급량 증가: 워케이션 관련 언급량이 지속적으로 상승하고 있으며, 관심도가 높아지고 있음.
2. 지역별 언급량 순위: 강원도 > 부산 = 서울 > 제주 > 경기 순으로 나타남.
3. 지역별 성장세 분석: 강원도와 부산은 꾸준한 성장세를 보이나, 경기·서울·제주는 성장 둔화 조짐.
4. 인기 지역 변화: 2021년에는 강원도, 2024년에는 제주도가 가장 많이 언급됨.
5. 관광 요소 비교: 워케이션 상위 30% 지역은 방문자 수, 관광 소비액, 숙박 방문자 비율이 높으나, 장기 숙박 비율이 낮음.
6. 숙박 형태와 소비 분석:
    - 단기 숙박 비율과 워케이션 언급량 간 유의미한 양의 관계 존재.
    - 장기 숙박 비율은 관광 소비액과 유의미한 양의 관계 존재.
    - 장기 체류형 워케이션이 지역 경제 활성화에 기여 가능.
7. 여행 유형 영향 분석: 자연 관광, 숙박이 워케이션 언급량에 영향을 미치는 주요 여행 유형.
8. 트렌드 영향 분석: 체험 및 기타 활동이 워케이션 언급량과 연관성이 높음.
9. 관광 지표 분석: 숙박과 관광 요소가 워케이션 언급량 상위 지역의 핵심 요소.
10. 최대 관광 지표: 자연 관광이 워케이션 언급량 상위 6개 지역의 주요 관광 지표.

## 3. 분석 결과 요약
(1) 워케이션 잠재 지역 선정 기준
- 워케이션 언급량은 높지만 방문자 수가 적은 지역
- 장기 숙박 비율이 높은 지역
- 장기 숙박 비율과 관광 소비액 간 상관관계가 개선되는 지역
- 선정된 상관관계 개선 지역과 공통된 지역 군집 선정
최종적으로 춘천시와 영도구가 워케이션 잠재 지역으로 선정되었다.

(2) 군집화 분석 결과: 군집 0 (저소비형)
- 자연관광 소비가 가장 높은 그룹이며 체험관광 소비도 상대적으로 높음.
- 가족 단위(부모-자녀) 방문이 많음
- 방문자 거리가 가장 길고, 체류시간이 가장 김.
- 멀리서 오는 장기 체류형 방문객이 많을 가능성이 있음
- 쇼핑 및 음식 소비가 상대적으로 낮음

(3) 워케이션 트렌드 지속 성장
- 워케이션 언급량은 지속적으로 상승 중이며, 특히 강원도, 부산, 서울, 제주, 경기 순으로 높은 관심을 받고 있음.
- 강원도와 부산은 지속적인 성장세를 보이지만, 서울·경기·제주는 성장률이 둔화되는 경향이 있음.

(4) 숙박 형태와 소비 패턴
- 워케이션 언급량이 높은 지역은 방문자 수, 관광 소비액, 숙박 방문자 비율이 높음.
- 그러나 장기숙박 비율은 상대적으로 낮으며, 단기숙박 비율이 워케이션 언급량과 양의 상관관계를 가짐.
- 반면 장기숙박 비율은 관광 소비액과 양의 상관관계를 가지므로, 체류시간 증가 시 지역 경제 활성화에 기여할 가능성이 큼.

(5) 주요 워케이션 요소
- 자연 관광 및 숙박이 워케이션과 밀접한 관련이 있으며, 체험 관광 또한 중요한 요소로 작용함.
- 가족 단위의 워케이션 방문자가 증가하고 있으며, 특히 프리랜서, 개인사업자 등 자율적 근무 형태를 가진 사람들이 선호함.     
- 지자체에서도 체류시간을 늘리기 위한 다양한 지원 정책을 추진 중.

## 4. 콘텐츠 마케팅 전략
(1) 콘텐츠 소재
- 가족형 워케이션의 이점을 강조한 영상 및 블로그 콘텐츠
- 춘천, 영도 지역의 자연관광·체험 프로그램을 소개하는 인스타그램/유튜브 콘텐츠
- 장기 체류의 장점 (업무 효율성, 가족과의 시간, 비용 절감 등)을 강조하는 카드뉴스 및 인터뷰
- 워케이션 숙소 및 공유 오피스 정보 제공
- 정부 지원 프로그램 및 혜택을 소개하는 콘텐츠 제작

(2) 페르소나 설정
- 페르소나 1: 프리랜서 & 개인사업자 
    - 35~45세, 디지털 노마드형 근무자
    - 일과 휴식 시간의 분리 중요
    - 업무 공간과 관광 요소 모두 고려
    - 여행지에서의 자녀 돌봄 서비스를 활용해 육아로부터의 자유를 경험하려는 니즈가 크다.
- 페르소나 2: 직장인 가족 (연차 활용형) :
    - 30~40대, 어린 자녀를 둔 부모
    - 연차를 활용하여 가족과 여행 겸 업무
    - 숙소, 관광지, 자녀와 함께하는 체험활동 중요
    - 직장 생활로 인한 피로를 해소하고, 자녀와의 시간을 더욱 확보하려는 목적이 강하다.
- 페르소나 3: 기업 워케이션 담당자
    - 기업 내 워케이션 제도 도입을 고려 중
    - 직원 복지를 고려하며, 업무 효율성 상승 기대
    - 지자체 지원 및 혜택 관심

(3) 콘텐츠 제작
- 블로그 및 브이로그: 워케이션 경험담, 지역 명소 소개, 체험활동 후기 등을 담은 콘텐츠를 제작하여 온라인에 공유합니다.
- 소셜 미디어 캠페인: 인스타그램, 페이스북 등을 활용한 사진 및 영상 콘텐츠로 지역의 매력을 홍보합니다.
- 온라인 세미나 및 웨비나: 워케이션의 장점과 지역별 프로그램을 소개하는 온라인 세미나를 개최하여 관심 있는 이들의 참여를 유도합니다.

## 6. STP 전략
- Segmentation (시장 세분화)
    - 가족 단위 워케이션 (원격 근무자, 디지털 노마드, 프리랜서)
    - 장기 체류형 워케이션 (장기 출장/ 프로젝트 근무자)
    - 기업형 워케이션 (직원복지/ 팀워크 향상 위한 프로젝트 팀·개인)
- Targeting (목표 시장 선정)
    - 가족 단위의 고객, 특히 아이들과 함께 자연·체험 활동을 원하는 프리랜서 혹은 개인 사업자 부모층
    - MZ세대 대상 기업 차원의 워케이션을 고려하는 중소기업 및 스타트업
- Positioning (포지셔닝)
    - (1) "가족과 함께하는 자연 체험, 일상 속 특별한 휴식과 추억을 만들어보세요."
        - 자연 속에서 아이들과 함께하는 체험 활동과 휴식을 결합하여 가족 단위 고객에게 특별한 경험을 제공.
        - 프리랜서나 개인 사업자 부모층의 업무와 휴가를 동시에 충족시키는 유연한 환경을 강조.
    
    - (2) "업무와 휴식의 완벽한 균형, MZ세대가 원하는 워케이션으로 기업 문화를 혁신하세요."
        - MZ세대 직원들이 선호하는 유연한 근무 환경과 휴식을 결합한 워케이션 프로그램 제공.
        - 중소기업 및 스타트업의 기업 문화 개선과 직원 만족도, 생산성 향상을 목표로 하는 메시지 전달.
    
## 7. 4P 전략
- Product (제품 전략)
    - 가족 맞춤형 워케이션 패키지 (숙박 + 체험 프로그램 + 공유 오피스)
    - 자녀 돌봄 서비스 및 지역 관광 상품권 제공
    - 자연 친화적 워케이션 숙소의 장기 숙박 상품 기획
- Price (가격 전략)
    - 장기 숙박 시 할인 혜택 제공
    - 워케이션 숙박객 대상 지역 상품권 증정
    - OTA(온라인 여행사) 플랫폼과 협업하여 패키지 할인 제공
- Place (유통 전략)
    - OTA 플랫폼(Yanolja, Agoda 등)과 협업하여 상품 노출 확대
    - 기업 대상 B2B 패키지 제공
    - 소셜미디어 및 디지털 광고 활용
- Promotion (프로모션 전략)
    - 가족 워케이션 팸투어 운영 (인플루언서 협업) 
    - 기업 HR 부서 및 인사 담당자 대상 이메일 마케팅
    - 지자체 지원 혜택 및 숙박 연계 콘텐츠 제작

## 8. 3C 전략
- Customer (고객)
    - 장기 숙박을 고려하는 가족 단위 방문객
    - 기업 워케이션을 도입하려는 인사 담당자
- Competitor (경쟁사)
    - 기존 워케이션 지역과 차별화 전략 필요
    - 해외 워케이션 거점(발리, 태국 치앙마이 등)
- Company (자사)
    - 지자체 및 OTA 협업을 통해 지역 특화 워케이션 상품 개발
    - 고객 맞춤형 장기 숙박 혜택 제공
    - 자연·체험 중심의 가족형 워케이션 특화 서비스 제공
